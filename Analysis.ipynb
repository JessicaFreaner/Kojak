```python
def get_pct_diff(xyear, yyear):
    return (yyear - xyear)/xyear
```

```python

def interpolate_values(df, mode='default'):
    df_list = []
    for i, nta in enumerate(df.NTACode):
        print nta
        if mode == 'default':
#         print nta
            df_2011 = pd.DataFrame(np.linspace(df.iloc[i]['adj_2010'], 
                                               df.iloc[i]['adj_2011'], 
                                               num =52))
            df_2012 = pd.DataFrame(np.linspace(df.iloc[i]['adj_2011'],
                                               df.iloc[i]['adj_2012'],
                                               num =52))
            df_2013 = pd.DataFrame(np.linspace(df.iloc[i]['adj_2012'], 
                                               df.iloc[i]['adj_2013'],
                                               num =52))
            df_2014 = pd.DataFrame(np.linspace(df.iloc[i]['adj_2013'],
                                               df.iloc[i]['adj_2014'],
                                               num =52))
            df_all = pd.concat([df_2011, df_2012, df_2013, df_2014])
        elif mode == 'pct_diff':
            df_2012 = pd.DataFrame(np.linspace(df.iloc[i]['adj_pct_diff_2011'],
                                               df.iloc[i]['adj_pct_diff_2012'],
                                               num =52))
            df_2013 = pd.DataFrame(np.linspace(df.iloc[i]['adj_pct_diff_2012'], 
                                               df.iloc[i]['adj_pct_diff_2013'],
                                               num =52))
            df_2014 = pd.DataFrame(np.linspace(df.iloc[i]['adj_pct_diff_2013'],
                                               df.iloc[i]['adj_pct_diff_2014'],
                                               num =52))
            df_all = pd.concat([df_2012, df_2013, df_2014])
            
        df_all.columns = [nta]
        df_all.reset_index()
        df_list.extend([df_all])
        
    return df_list
```

```python
def plot_yr_prop_values_and_prev_yr(df, xyear, yyear, mode='default', all_yrs=False):

    X = df[[xyear, 'BK','BX','MN','QN','SI']]
    if all_yrs:
        X = df[['adj_2010', 'adj_2011', 'adj_2012', 'adj_2013','BK','BX','MN','QN','SI']]
    X = sm.add_constant(X)
    if mode == 'default':
        Y = df[yyear] 
    elif mode == 'diff':
        Y = df[yyear] - df[xyear]
    elif mode == 'pct_diff':
        X = df[['BK','BX','MN','QN','SI']]
        X = sm.add_constant(X)
        
        Y = ( df[yyear] - df[xyear] ) / df[xyear]
    elif mode == 'log':
        X =  np.log(df[[xyear]])
        X = sm.add_constant(X)
        Y = np.log(df[yyear])
 
    linmodel = sm.OLS(Y, X).fit()

    plt.scatter(df[xyear], Y)
    plt.xlabel('%s Median Prices' % xyear)
    plt.ylabel('%s Median Prices' % yyear)
    plt.title("Prices Based on Last Year's Prices")
    if mode == 'default':
        plt.plot(df[xyear], linmodel.predict(), 'r')
    elif mode == 'diff':
        plt.plot(df[xyear], linmodel.predict(), 'c')
    elif mode == 'pct_diff':
#         plt.plot(df['xyear'], linmodel.predict(), 'm')
        plt.hist( (df[yyear] - df[xyear] ) / df[xyear], bins=20)

    elif mode == 'log':
        plt.plot(df[xyear], linmodel.predict(), 'g')
```

# -----OUTLIERS-----


## Drop above 1.5 * IQR + Median


```python
def find_interquartile_range(x):
    q75, q25 = np.percentile(x, [80 ,20])
    return q75 - q25



def drop_outliers(df, count='entry', limit='default', iqr_val=10):
    if count == 'entry':
        period_cnt = 'entry_period_cnt'
        if limit == 'iqr':
            iqr = find_interquartile_range(df[period_cnt].values)
            top_limit = df[period_cnt].median() + iqr * iqr_val
        else:
            top_limit = 1000000
        print '\nENTRIES:'
    elif count == 'exit':
        period_cnt = 'exit_period_cnt'
        if limit == 'iqr':
            iqr = find_interquartile_range(df[period_cnt].values)
            top_limit = df[period_cnt].median() + iqr * iqr_val
        else:
            top_limit = 1000000
        print '\nEXITS:'
    
    drop_cnt = len(df[df[period_cnt] >= top_limit])
    total_cnt = len(df[df[period_cnt] < top_limit])
    drop_df = df[df[period_cnt] >= top_limit]
    return_df = df[df[period_cnt] < top_limit]

        
    print '\tNum records dropped:\t', drop_cnt, '\t\tCut off:\t', top_limit
    print '\tNum records retained:\t', total_cnt
    
    return drop_df, return_df
```
