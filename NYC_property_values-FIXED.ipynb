```python
def make_property_value_dict(df):
    pv_dict = {}
    for i, index in enumerate(df.index):
        key = index
        values = {}
        for col in df.columns[:-1]:
            values[col] = df.iloc[i][col]
        pv_dict[index] = values
#         print pv_dict
#         break
    return pv_dict
```

```python
bk_med_val_dict = make_property_value_dict(grouped_median_bk_nta_df)
manh_med_val_dict = make_property_value_dict(grouped_median_manh_nta_df)
qu_med_val_dict = make_property_value_dict(grouped_median_qu_nta_df)
bx_med_val_dict = make_property_value_dict(grouped_median_bx_nta_df)
st_med_val_dict = make_property_value_dict(grouped_median_st_nta_df)
```

```python
def make_adj_property_value_dict(df):
    pv_dict = {}
    for i, index in enumerate(df.index):
        key = index
        values = {}
        for col in df.columns[:]:
            values[col] = df.iloc[i][col]
        pv_dict[index] = values
#         print pv_dict
#         break
    return pv_dict
```

```python
def add_prop_values_to_json(json_dict, prop_val_dict):
    for i, nta in enumerate(json_dict['features']):
#         print nta['properties']['NTAName']
        nta_name = nta['properties']['NTAName']
        try:
            nta['properties']['PropValues'] = prop_val_dict[nta_name]
        except:
            print "Failed to find PROPERTY VALUES for %s" % nta_name
```

```python
add_prop_values_to_json(data_bk, bk_med_val_dict)
add_prop_values_to_json(data_manh, manh_med_val_dict)
add_prop_values_to_json(data_qu, qu_med_val_dict)
add_prop_values_to_json(data_bx, bx_med_val_dict)
add_prop_values_to_json(data_st, st_med_val_dict)
```

```python
def add_adj_prop_values_to_json(json_dict, prop_val_dict):
    for i, nta in enumerate(json_dict['features']):
#         print nta['properties']['NTAName']
        nta_code = nta['properties']['NTACode']
        nta_name = nta['properties']['NTAName']
        try:
            nta['properties']['AdjPropValues'] = prop_val_dict[nta_code]
        except:
            print "Failed to find PROPERTY VALUES for %s" % nta_name
```
